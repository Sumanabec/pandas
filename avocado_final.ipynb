```python
import pandas as pd
```

```python
def read_and_filter_data(data_file, separator, relevant_category_file):
    df = pd.read_csv(data_file, sep=separator)
    
    relevant_columns = ['code', 'lc', 'product_name_en', 'quantity', 'serving_size', 'packaging_tags', 'brands', 'brands_tags', 'categories_tags', 'labels_tags', 'countries', 'countries_tags', 'origins', 'origins_tags']
    df_cleaned = df[relevant_columns]

    #processing the dataframe
    df_cleaned.dropna(subset=['categories_tags'], inplace = True)
    df_cleaned['categories_tags_list']= df_cleaned['categories_tags'].str.split(',')

    #read the relevant tags
    avocado_tags=[]
    with open(relevant_category_file, 'r') as file:
        avocado_tag_list = file.readlines()
    avocado_tags = [tag.strip() for tag in avocado_tag_list]

    #return the dataframes which contains above tags
    df_cleaned = df_cleaned[df_cleaned['categories_tags_list'].apply(lambda x: any([tag for tag in x if tag in avocado_tags]) )]

    df_cleaned['countries_list'] = df_cleaned['countries'].str.split(',')
    df_avocado = df_cleaned.explode('countries_list')

    df_uk = df_avocado[df_avocado['countries_list'] == 'United Kingdom']
    df_uk['origin_tags_list'] = df_uk['origins_tags'].str.split(',')
    df_uk = df_uk.explode('origin_tags_list')

    top_origin_string = (df_uk['origin_tags_list'].value_counts().index[0])
    top_origin_country = top_origin_string.lstrip('en:').title()

    return top_origin_country
    
    
    
    
    
```

```python
top_avocado_origin = read_and_filter_data('avocado.csv',',','relevant_avocado_categories.txt')
top_avocado_origin
```
